# Question 3

Next, we need to check the distribution of the duration variable. There are some outliers. Let's remove them and keep only the records where the duration was between 1 and 60 minutes (inclusive).

What fraction of the records left after you dropped the outliers?

```python
total_records = df_train.shape[0]
total_records
```

```python
df_train.dtypes
```

```python
df_train = df_train[(df_train["duration"] >= 1) & (df_train["duration"] <= 60)]
df_train
```

```python
frac_no_outliers = (total_filtered * 100) / total_records
frac_no_outliers
```

# Question 4 
Let's apply one-hot encoding to the pickup and dropoff location IDs. We'll use only these two features for our model.

    - Turn the dataframe into a list of dictionaries (remember to re-cast the ids to strings - otherwise it will label encode them)
    - Fit a dictionary vectorizer
    - Get a feature matrix from it

What's the dimensionality of this matrix (number of columns)?

```python
from sklearn.feature_extraction import DictVectorizer 
```

```python
FEATURES_CAT = ['PULocationID','DOLocationID']
```

```python
df_train[FEATURES_CAT] = df_train[FEATURES_CAT].astype(str)
```

```python
df_train_vec = df_train[FEATURES_CAT].to_dict(orient='records')
df_train_vec
```

```python
dv = DictVectorizer()
dv.fit_transform(df_train_vec)
```

## Answer: 515

# Question 5
Now let's use the feature matrix from the previous step to train a model.

    Train a plain linear regression model with default parameters, where duration is the response variable
    Calculate the RMSE of the model on the training data

What's the RMSE on train?

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error
```

```python
X_train = dv.fit_transform(df_train_vec)
y_train = df_train["duration"].values

lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_train)
y_pred
```

```python
score = root_mean_squared_error(y_train,y_pred)
score
```

## Answer: 7.64

# Question 6
Now let's apply this model to the validation dataset (February 2023).

What's the RMSE on validation?

For the cleaning of the validation dataset, i will use the following function that compiles all the steps of the training set:

```python
def read_dataframe(filename):
    df = pd.read_parquet(filename)
    df["duration"] = df_train.tpep_dropoff_datetime - df_train.tpep_pickup_datetime
    df.duration = df.duration.apply(lambda tm: tm.total_seconds() / 60)
    df = df[(df.duration >= 1) & (df.duration <= 60)]
    categorical = ['PULocationID', 'DOLocationID']
    df[categorical] = df[categorical].astype(str)
    return df
```

```python
df_validate = read_dataframe("data/yellow_tripdata_2023-02.parquet")
df_validate
```

```python
y_val = df_validate["duration"].values
```

```python
df_val_vec = df_validate[FEATURES_CAT].to_dict(orient='records')
len(df_val_vec)
```

```python
X_val = dv.transform(df_val_vec)
```

```python
y_val_pred = lr.predict(X_val)
```

```python
score_val = root_mean_squared_error(y_val,y_val_pred)
score_val
```

## Answer 11.81
